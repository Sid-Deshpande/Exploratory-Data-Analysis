```python
# we want country as well for the sake of getting information about ciies
# so that first we gather dataframe of country

country_df = pd.read_excel('/content/Country-Code.xlsx')
```

```python
# now we merge this two dataframe as single

final_df = pd.merge(zomato_df,country_df,on='Country Code', how = 'left')
final_df.sample(2)
```

```python
list(final_df['Cuisines'].unique())
```

```python
final_df.fillna({'Cuisines':'unknown'},inplace = True)
```

```python
# now which city got the max votes and what rating

max_votes_by_city = final_df.groupby(['City','Aggregate rating','Rating color'])['Votes'].max().reset_index().sort_values(by ='Votes', ascending = False)
max_votes_by_city.head(10)
```

```python


# Sort the DataFrame by maximum votes in descending order
max_votes_by_city_sorted = max_votes_by_city.sort_values(by='Votes', ascending=False)

# Select the top 10 cities
top_10_cities = max_votes_by_city_sorted.head(10)

# Set the style
sns.set(style="whitegrid")

# Create the bar plot
plt.figure(figsize=(12, 8))
sns.barplot(data=top_10_cities, x='City', y='Votes', hue='Rating color')
plt.title('Top 10 Cities by Maximum Votes with Rating Color')
plt.xlabel('City')
plt.ylabel('Maximum Votes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# now hotel name and the city according to the aggregate rating

famous_hotel_by_rating = final_df.groupby(['Restaurant Name','Address','City','Country'])['Aggregate rating'].max().reset_index().sort_values(by = 'Aggregate rating', ascending = False)
famous_hotel_by_rating.head(20)
```

```python

# Sort the DataFrame by aggregate rating in descending order
famous_hotels_sorted = famous_hotel_by_rating.sort_values(by='Aggregate rating', ascending=False)

# Select the top 10 famous hotels
top_10_hotels = famous_hotels_sorted.head(10)

# Set the style
sns.set(style="whitegrid")

# Create the bar plot
plt.figure(figsize=(12, 8))
sns.barplot(data=top_10_hotels, x='Restaurant Name', y='Aggregate rating')
plt.title('Top 10 Famous Hotels by Rating')
plt.xlabel('Hotel Name')
plt.ylabel('Aggregate Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# finding name of the restaurants where city = banglore

# Filter the DataFrame for restaurants in Bangalore
bangalore_restaurants = final_df[final_df['City'] == 'Bangalore']

# Extract the names of the restaurants
bangalore_restaurant_names = bangalore_restaurants['Restaurant Name'].unique()

# Print the names of the restaurants in Bangalore
for restaurant_name in bangalore_restaurant_names:
    print(restaurant_name)
```

```python
# Filter the DataFrame for restaurants in Bangalore
bangalore_restaurants = final_df[final_df['City'] == 'Bangalore']

# Extract the names, aggregate ratings, and text reviews of the restaurants
bangalore_restaurant_info = bangalore_restaurants[['Restaurant Name','Price range','Aggregate rating','Rating text']]

# Remove duplicate rows to avoid redundancy
bangalore_restaurant_info.drop_duplicates(inplace=True)

# Print the information of restaurants in Bangalore
print(bangalore_restaurant_info)
```

```python


# Filter the DataFrame for restaurants in Bangalore
bangalore_restaurants = final_df[final_df['City'] == 'Bangalore']

# Extract the names, aggregate ratings, and text reviews of the restaurants
bangalore_restaurant_info = bangalore_restaurants[['Restaurant Name', 'Aggregate rating']]

# Remove duplicate rows to avoid redundancy
bangalore_restaurant_info.drop_duplicates(inplace=True)

# Sort the DataFrame by aggregate rating
bangalore_restaurant_info.sort_values(by='Aggregate rating', ascending=False, inplace=True)

# Visualize the aggregate ratings of restaurants in Bangalore
plt.figure(figsize=(10, 6))
plt.barh(bangalore_restaurant_info['Restaurant Name'][:10], bangalore_restaurant_info['Aggregate rating'][:10])
plt.xlabel('Aggregate Rating')
plt.ylabel('Restaurant Name')
plt.title('Top 10 Restaurants in Bangalore by Aggregate Rating')
plt.gca().invert_yaxis()  # Invert y-axis to display highest rating at the top
plt.show()
```

Heatmap of Average Cost for two vs. Price range:

```python
# Selecting only numerical columns
numerical_columns = final_df.select_dtypes(include=['int64', 'float64'])

# Calculating the correlation matrix
correlation_matrix = numerical_columns.corr()

# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# which cuisines is the best in the world or we can say that which is the most ordered cuisines in the world

# Extracting the required columns from final_df
restaurant_data = {
    'Restaurant Name': final_df['Restaurant Name'],
    'Cuisines': final_df['Cuisines']
}

# Create a DataFrame from the extracted data
df_restaurants = pd.DataFrame(restaurant_data)

# Split cuisines into individual entries and count occurrences
cuisine_counts = df_restaurants['Cuisines'].str.split(', ', expand=True).stack().value_counts()

# Display the counts of each cuisine
print(cuisine_counts)
```

```python
# Plot the top cuisines by order count
top_cuisines = cuisine_counts.head(10)
top_cuisines.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Top 10 Most Ordered Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Order Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
